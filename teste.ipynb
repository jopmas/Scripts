```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections as mcollections

# Gerar dados
t = np.linspace(0, 10, 100)  # Série temporal
x = t
y = 2 * t  # Função linear

# Criar segmentos da linha
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

# Criar uma LineCollection com um colormap baseado na série temporal t
norm = plt.Normalize(t.min(), t.max())
lc = mcollections.LineCollection(segments, cmap='viridis', norm=norm)
lc.set_array(t)

# Configurar o plot
fig, ax = plt.subplots()
ax.add_collection(lc)
ax.autoscale()
ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())

# Adicionar uma barra de cores
plt.colorbar(lc, ax=ax, label='Tempo')

# Mostrar o plot
plt.show()
```
